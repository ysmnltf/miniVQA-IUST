In this notebook, a **VQA** model is implemented using **PyTorch** library.

- Question features are extracted using
  - **DistilBert** (no fine-tuning)
- Image features are available in the dataset.
- The question and image features are fused with
  - **Cross-attention** (with Multihead attention)
- The correct answer is predicted with a Dense layer.

**Best Validation Accuracy: 0.936**

# Build model

```python
# empty memory
del tokenizer
del image_features
del questions
import gc
gc.collect
```

```python
nhead = 4
```

```python
class MiniVQA(nn.Module):
    def __init__(self, text_size, image_size, nhead=nhead, dim_feedforward=2048, dropout=0.2, layer_norm_eps=1e-5):
        super(type(self), self).__init__()
        self.bert = DistilBertModel.from_pretrained('distilbert-base-uncased')
        self.cross_attn = nn.MultiheadAttention(
            embed_dim = text_size,
            num_heads = nhead,
            kdim = image_size,
            vdim = image_size,
            batch_first=True) # l:seq length n:batch size e:embed size
 
        # Implementation of Feedforward model
        self.linear1 = nn.Linear(image_size, dim_feedforward)
        self.norm1 = nn.LayerNorm(text_size, eps=layer_norm_eps)
        self.dropout1 = nn.Dropout(dropout)

        self.dec_norm = nn.BatchNorm1d(text_size * 17)
        self.dec_linears = nn.Sequential(
            nn.Linear(text_size * 17, 10)
        )
            
    def forward(self, text_features, image):
        
        text = self.bert(input_ids=text_features).last_hidden_state
        src2 = self.cross_attn(text, image, image)[0]
        src = text + self.dropout1(src2)
        src = self.norm1(src)
        output = torch.flatten(src, start_dim=1)
        output = self.dec_norm(output)
        output = self.dec_linears(output)

        return nn.functional.softmax(output, dim=1)
```

```python
miniVQA = MiniVQA(768, 512//nhead)
miniVQA.to(device)
```

# Train model

## Define constants

```python
learning_rate = 1e-4
epochs = 5
```

## Define train loop

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(miniVQA.parameters(), lr=learning_rate)
```

```python
def pred_val(model, dataloader):
  size = len(dataloader.dataset)
  correct = 0
  avg_loss = 0
  for batch, (text, image, y) in enumerate(dataloader):
    image_reshaped = torch.reshape(image, (image.shape[0], nhead, image.shape[1]//nhead))
    pred = model(text.long().to(device), image_reshaped.to(device))
    loss = loss_fn(pred, y.to(device))
    output = [torch.argmax(o).item() for o in pred]
    correct += (torch.FloatTensor(output) == y).float().sum()
    avg_loss += loss.item()
  acc = correct / size
  return avg_loss, correct, acc
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    correct = 0
    avg_loss = 0
    for batch, (text, image, y) in enumerate(dataloader):  
        # Compute prediction and loss
        image_reshaped = torch.reshape(image, (image.shape[0], nhead, image.shape[1]//nhead))
        pred = model(text.long().to(device), image_reshaped.to(device))
        loss = loss_fn(pred, y.to(device))
        # Backpropagation
        optimizer.zero_grad()
        loss.backward(retain_graph=True)
        optimizer.step()
        
        output = [torch.argmax(o).item() for o in pred]
        correct += (torch.FloatTensor(output) == y).float().sum()
        avg_loss += loss.item()

    avg_loss /= (size // 128 + 1)
    acc = correct / size
    val_loss, val_correct, val_acc = pred_val(miniVQA, valid_dataloader)
    print(f"training/ loss: {avg_loss:>7f} | accuracy: {acc}")
    print(f"val/ loss: {val_loss:>7f} | accuracy: {val_acc}")
```

```python
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, miniVQA, loss_fn, optimizer)
print("Done!")
```

# Predict
